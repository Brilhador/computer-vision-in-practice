### 1. Use o seu conhecimento aprendido nas lições anteriores e escreva um programa que conte a quantidade de moedas presentes na imagem moedas.jpg. 

Dica: use a função "findContours" disponibilizada pela biblioteca OpenCV. Referência: https://docs.opencv.org/trunk/d4/d73/tutorial_py_contours_begin.html

```python
### importação da biblioteca opencv, matplotlib, numpy
import cv2
import matplotlib.pyplot as plt
import numpy as np

# abrindo a imagem
imagem = cv2.imread('../../figures/moedas.jpg')
# converte imagem para modelo de cor RGB
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# mostrando a imagem do ruido
plt.imshow(imagem)
```

```python
# converterndo a imagem para tons de cinza
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# aplicando a suavização da imagem com uma mascara 7x7
imagem_suavizada = cv2.blur(imagem_gray, (7,7))

# gerando a imagem binarizada com limiar adaptativo
ret,imagem_binarizada = cv2.threshold(imagem_suavizada,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# aplicando o detector de bordas para identificar os objetos
imagem_bordas = cv2.Canny(imagem_binarizada,100, 200)

# mostrando a imagem do ruido
plt.imshow(imagem_bordas, cmap='gray')
```

#### Aplicando a função "findContours" para contar o número de contornos externos e encontrar a quantidade de moedas.

```python
# conta apenas os cortornos externos
objetos, lx = cv2.findContours(imagem_bordas, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
#A variável lx (lixo) recebe dados que não são utilizados

# desenhando os contornos 
# -1 significa desenha todos os contornos
cv2.drawContours(imagem, objetos, -1, (255, 0, 0), 3) 

# mostra a quantidade de moedas
print("Quantidade de moedas é:", len(objetos))

# mostrando a imagem do ruido
plt.imshow(imagem)
```

### 2. Baseado na solução anterior escreva um programa para contar a quantidade de dados presentes na imagem dados.jpg.


```python
### importação da biblioteca opencv, matplotlib, numpy
import cv2
import matplotlib.pyplot as plt
import numpy as np

# abrindo a imagem
imagem = cv2.imread('../../figures/dados.jpg')
# converte imagem para modelo de cor RGB
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# mostrando a imagem do ruido
plt.imshow(imagem)
```

```python
# abrindo a imagem
imagem = cv2.imread('../../figures/dados.jpg')

# converterndo a imagem para tons de cinza
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# aplicando a suavização da imagem com uma mascara 15x15
imagem_suavizada = cv2.GaussianBlur(imagem_gray, (15,15), 0)

# segmenta os dados da imagem
ret, imagem_binarizada = cv2.threshold(imagem_suavizada, 80, 255, cv2.THRESH_BINARY)

# # aplicando o detector de bordas para identificar os objetos
imagem_bordas = cv2.Canny(imagem_binarizada, 100, 200)

# mostrando a imagem do ruido
plt.imshow(imagem_bordas, cmap='gray')
```

```python
# conta apenas os cortornos externos
objetos, lx = cv2.findContours(imagem_bordas, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
#A variável lx (lixo) recebe dados que não são utilizados

# desenhando os contornos 
# -1 significa desenha todos os contornos
cv2.drawContours(imagem, objetos, -1, (255, 0, 0), 3) 

# exibe a quantidade de dados
print("Quantidade de dados é:", len(objetos))

# mostrando a imagem do ruido
plt.imshow(imagem)
```
