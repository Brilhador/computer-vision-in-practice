### 1. Aplique um dos métodos de binarização visto na lição sobre a imagem bandeiras.jpg de forma que somentes a corda e as bandeiras vermelhas permaneçam na imagem.

```python
### importação da biblioteca opencv, matplotlib, numpy
import cv2
import matplotlib.pyplot as plt
import numpy as np

# abrindo a imagem
imagem = cv2.imread('../../figures/bandeiras.jpg')
# converte imagem para modelo de cor RGB
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# mostrando a imagem do ruido
plt.imshow(imagem)
```

```python
# limiarizando a imagem com o valor de corte igual 100
ret, imagem_binarizada = cv2.threshold(imagem, 100, 255,cv2.THRESH_BINARY)
# mostrando a imagem binarizada
plt.imshow(imagem_binarizada)
```

### 2. Aplique um dos métodos de binarização visto na lição sobre a imagem bandeiras.jpg de forma que somentes a corda permaneça na imagem.

```python
# abrindo a imagem
imagem = cv2.imread('../../figures/bandeiras.jpg')
# converte imagem para tons de cinza
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# limiarizando a imagem com o valor de corte igual 50
ret, imagem_binarizada = cv2.threshold(imagem, 50, 255,cv2.THRESH_BINARY)
# mostrando a imagem binarizada
plt.imshow(imagem_binarizada, cmap="gray")
```
